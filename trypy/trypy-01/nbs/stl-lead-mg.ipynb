Making a new variable that contains the link to the csv file.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
from scipy.stats import norm
```

```python
url = 'https://raw.githubusercontent.com/'\
'allisonhorst/EDS_221_programming-essentials/'\
'main/activities/stl_blood_lead.csv'
```

pandas can read a csv file into a datagrame directly from a url:


```python
stl_lead = pd.read_csv(url)
stl_lead.describe
```

```python
stl_lead['popWhite'] = stl_lead['white']/stl_lead['totalPop'] 
```

```python
#making a scatterplot
x = stl_lead['popWhite']
y = stl_lead['pctElevated']

plt.scatter(x,y, c= y, cmap = "viridis", marker ='D')
```

Making a histogram


```python
x = stl_lead['pctElevated']
plt.hist(x, bins='auto', color='#b434eb',
                            alpha=0.4, rwidth=0.85, edgecolor = "black")
plt.xlabel("Percent of elevated blood lead level ")
plt.ylabel("Count")
plt.title("Histogram of Percent elevated bood level ")

```
